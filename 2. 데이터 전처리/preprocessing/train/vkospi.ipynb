```python
import pandas as pd
import numpy as np
```

```python
vkospi = pd.read_csv('../../../1. 데이터셋/train_data/VKOSPI.csv', index_col=0)
```

```python
vkospi = vkospi[::-1]
```

```python
vkospi = vkospi.reset_index()
vkospi
```

```python
vkospi['당일 변동율'] = round((vkospi['종가'] - vkospi['시가']) / vkospi['시가'] * 100, 3)
```

```python
new_label = []
for i in range(len(vkospi['날짜'])):
	if vkospi['당일 변동율'][i] > 1:
		result = 1
		new_label.append(result)
	else:
		result = 0
		new_label.append(result)
```

```python
label = []
for i in range(len(vkospi['날짜'])):
	if vkospi['시가'][i] > vkospi['종가'][i]:
		result = 0
		label.append(result)
	else:
		result = 1
		label.append(result)
```

```python
new_label_df = pd.DataFrame(new_label, columns=["VKOSPI_new_Label"])
label_df = pd.DataFrame(label, columns=["VKOSPI_Label"])
```

```python
vkospi = pd.concat([vkospi, label_df], axis=1)
vkospi = pd.concat([vkospi, new_label_df], axis=1)
```

```python
vkospi['날짜'] = pd.to_datetime(vkospi['날짜'])
```
